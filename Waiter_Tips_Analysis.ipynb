# WAITER TIPS ANALYSIS 

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Waiter_Tips.csv")
df
```

## 1.What is the average tip amount given by customers? 

```python
average_tip = df['tip'].mean()
print("The average tip amount in dollars ($) is:", average_tip)
```

## 2.How many transactions were made by male customers vs. female customers?

```python
transactions_by_gender = df['sex'].value_counts()
print(transactions_by_gender)
```

```python
gender_percentages = df['sex'].value_counts(normalize=True) * 100

# Create a pie chart
plt.pie(gender_percentages, labels=gender_percentages.index, autopct='%1.1f%%')

# Add a title
plt.title('Percentage of Transactions by Gender')

# Show the chart
plt.show()
```

## 3.What is the total bill amount for each day of the week?

```python
total_bills_by_day = df.groupby('day')['total_bill'].sum()
print(total_bills_by_day)
```

```python
total_bills_by_day = df.groupby('day')['total_bill'].sum()

for x, y in zip(total_bills_by_day.index, total_bills_by_day.values):
    plt.text(x, y, f"{y:.2f}", ha='center', va='bottom', fontweight='bold')
    
# Create a bar chart
plt.bar(total_bills_by_day.index, total_bills_by_day.values)

# Add axis labels and a title
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill Amount in dollars ($)')
plt.title('Total Bill Amount by Day in dollars ($)')

# Show the chart
plt.show()
```

## 4.What is the total tip amount for each day of the week?

```python
total_tips_by_day = df.groupby('day')['tip'].sum()
print(total_tips_by_day)
```

```python
total_tips_by_day = df.groupby('day')['tip'].sum()

for x, y in zip(total_tips_by_day.index, total_tips_by_day.values):
    plt.text(x, y, f"{y:.2f}", ha='center', va='bottom', fontweight='bold')

# Create a bar chart
plt.bar(total_tips_by_day.index, total_tips_by_day.values)

# Add axis labels and a title
plt.xlabel('Day of the Week')
plt.ylabel('Total Tip Amount in dollars($)')
plt.title('Total Tip Amount by Day in dollars ($)')

# Show the chart
plt.show()
```

## 5.How many transactions were made by smokers vs. non-smokers?

```python
transactions_by_smoker = df.groupby('smoker')['total_bill'].count()
print(transactions_by_smoker)
```

```python
transactions_by_smoker = df.groupby('smoker')['total_bill'].count()

# Create a pie chart
plt.pie(transactions_by_smoker.values, labels=transactions_by_smoker.index, autopct='%1.1f%%')

# Add a title
plt.title('Proportion of Transactions by Smoker')

# Show the chart
plt.show()
```

## 6.What is the average total bill amount for parties of different sizes?

```python
avg_bill_by_size = df.groupby('size')['total_bill'].mean()
print(avg_bill_by_size)
```

```python
avg_bill_by_size = df.groupby('size')['total_bill'].mean()

for x, y in zip(avg_bill_by_size.index, avg_bill_by_size.values):
    plt.text(x, y, f"{y:.2f}", ha='center', va='bottom', fontweight='bold')

# Create a bar graph with yellow bars
plt.bar(avg_bill_by_size.index, avg_bill_by_size.values, color='Brown')

# Set the x-axis label
plt.xlabel('Party Size')

# Set the y-axis label
plt.ylabel('Average Total Bill in dollars ($)')

# Set the title
plt.title('Average Total Bill in dollars ($) by Party Size')

# Show the chart
plt.show()
```

## 9.What is the average tip amount for lunch vs. dinner transactions?

```python
tip_by_time = df.groupby('time')['tip'].mean()

# Print the result
print(tip_by_time)
```

```python
tip_by_time = df.groupby('time')['tip'].mean()

# Create a bar chart
plt.bar(tip_by_time.index, tip_by_time.values, color='blue')

for x, y in zip(tip_by_time.index, tip_by_time.values):
    plt.text(x, y, f"{y:.2f}", ha='center', va='bottom', fontweight='bold')

# Set the title and axis labels
plt.title('Average Tip Amount in dollars ($) by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Average Tip Amount in dollars ($)')
plt.grid()

# Show the chart
plt.show()
```

## 10.What is the average tip percentage (tip amount as a percentage of total bill) for each day of the week?

```python
df['tip_pct'] = df['tip'] / df['total_bill'] * 100

# Group the data by day of the week and calculate the mean of tip_pct for each group
tip_pct_by_day = df.groupby('day')['tip_pct'].mean()

# Print the result
print(tip_pct_by_day)
```

```python
plt.bar(tip_pct_by_day.index, tip_pct_by_day.values, color='green')

for x, y in zip(tip_pct_by_day.index, tip_pct_by_day.values):
    plt.text(x, y, f"{y:.2f}", ha='center', va='bottom', fontweight='bold')


# Set the title and axis labels
plt.title('Average Tip Percentage by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Average Tip Percentage')

# Show the chart
plt.show()
```
